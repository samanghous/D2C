```python
#*******************************IMPORTING DATASET AND STRUCTURING DATA********************************************
```

```python
import pandas as pd

full_data=pd.read_excel('Training Dataset.xlsx')
partial_data=pd.read_excel('Testing Dataset.xlsx')

def classify(x):
    if(x=='GALAXY'):
        return 0
    elif(x=='STAR'):
        return 1
    elif(x=='QSO'):
        return 2

full_data['class_GSQ']=full_data['class'].apply(classify)    
```

```python
#********************************************FULL DATA***********************************************************
```

```python
full_data
```

```python
full_data1=full_data.drop(['rerun','objid','run','camcol'],axis=1)
full_data1                                                                     # WITH SOME FEATURES REMOVED
```

```python
full_data1['ra']= full_data1['ra']*1000
full_data1['dec']= full_data1['dec']*10000
full_data1['u']= full_data1['u']*10
full_data1['g']= full_data1['g']*10
full_data1['r']= full_data1['r']*1000
full_data1['i']= full_data1['i']*1000
full_data1['z']= full_data1['z']*1000
full_data1['redshift']= full_data1['redshift']*10000000
full_data1['specobjid']= full_data1['specobjid']/10000000000

```

```python
#********************************FULL DATA 1 AND ITS DIVISION****************************************************
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(full_data1.iloc[:,0:13],full_data1['class_GSQ'],test_size=0.1)
```

```python
#***********************************FULL DATA 1 TRAINING SCALE UP*************************************************
                               # AS WELL WHOLE FULL DATA 1 SCALE UP
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=27 )
x_train_new,y_train_new= sm.fit_sample(x_train,y_train)

sm = SMOTE(random_state=27 )
x_fulldata1,y_fulldata1= sm.fit_sample(full_data1.iloc[:,0:13],full_data1['class_GSQ'])
```

```python
#***********************************HYPERPARAMETER TUNING THINGS************************************************
```

```python
from sklearn.model_selection import RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier
sammodel=RandomForestClassifier(n_jobs=-1)
```

```python
sam_hyperparam={
    'n_estimators':[75,80,85,87,90,95],
    'criterion':['entropy'],
    'max_depth':[None,80,82,84,85,87,90],
    'min_samples_split':[3,4,5,6,7,8,9],
    'min_samples_leaf':[3],
    'bootstrap':[False],
    'max_features':['auto','sqrt','log2',4,5,6,7,8,9],
    'class_weight':["balanced_subsample"]
}



def sam_hyperpm_func(itern):
    from sklearn.model_selection import RandomizedSearchCV
    sam_bestparam=RandomizedSearchCV(sammodel,param_distributions=sam_hyperparam,n_jobs=-1,n_iter=itern,cv=5)
    sam_bestparam.fit(x_fulldata1,y_fulldata1)
    gh1=sam_bestparam.best_params_ 
    gh2=sam_bestparam.best_score_
    return gh1,gh2
```
