```python
#*******************************IMPORTING DATASET AND STRUCTURING DATA********************************************
```

```python
import pandas as pd
```

```python
full_data=pd.read_excel('Training Dataset.xlsx')
partial_data=pd.read_excel('Testing Dataset.xlsx')
```

```python
full_data
```

```python
partial_data
```

```python
def classify(x):
    if(x=='GALAXY'):
        return 0
    elif(x=='STAR'):
        return 1
    elif(x=='QSO'):
        return 2

full_data['class_GSQ']=full_data['class'].apply(classify)    
```

```python
full_data
```

```python
full_data.shape
```

```python
# ***************WE CAN REMOVE A SMALL FRACTION FROM TRAINIG DATA FOR TESTING PURPOSE******************************
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(full_data.iloc[:,0:17],full_data['class_GSQ'],test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
#***********************IMPORT AND APPLY RANDOM FOREST ON SLICED TRAINING*****************************************
```

```python
from sklearn.ensemble import RandomForestClassifier
sammodel=RandomForestClassifier()
```

```python
sammodel.fit(x_train,y_train)
```

```python
#**********************************SCORE OF SLICED TESTING DATA****************************************************
```

```python
sammodel.score(x_test,y_test)
```

```python
ytest_pred = sammodel.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, recall_score
```

```python
accuracy_score(y_test,ytest_pred)
```

```python
 y_test
```

```python
#******************************SEE WHICH FEATURE HAS LEAST EFFECT ON TRAINING**************************************
```

```python
pd.Series(sammodel.feature_importances_,index=full_data.columns[0:17]).sort_values(ascending=False)
```

```python
#**************************** REMOVING LEAST IMPORTANT FEATURE AND TARGET ***************************************
```

```python
full_data.drop(['camcol','run','rerun','objid','field','fiberid','class','class_GSQ'],axis=1)
```

```python
#******************************AGAIN APPLY RANDOM FOREST WITH LESS FEATURES***************************************
```

```python
sammodel=RandomForestClassifier(n_estimators=600)
```

```python
sammodel.fit(x_train.drop(['camcol','run','rerun','objid','field','fiberid'  ],axis=1) ,y_train)
```

```python
#***********************************SCORE PLEASE!!!**************************************************************
```

```python
sammodel.score(x_test.drop(['camcol','run','rerun','objid','field','fiberid'  ],axis=1),y_test)
```

```python
ytest_pred = sammodel.predict(x_test.drop(['camcol','run','rerun','objid','field','fiberid'  ],axis=1))
```

```python
from sklearn.metrics import confusion_matrix
sam_cm=confusion_matrix(y_test,ytest_pred)
sam_cm
```

```python
#***************************************HERE TRAINING IS DONE!!!**************************************************
```

```python
#***********************************USE TRAINED MODEL TO FILL PARTIAL DATA****************************************
```

```python
predicted_class_GSQ=sammodel.predict(partial_data.drop(['camcol','run','rerun','objid','field','fiberid','class'],axis=1))
predicted_class_GSQ
```

```python
#***************************************MAKING PROPER RESULT DATA*************************************************
```

```python
predicted_class=[]
for i in range(len(predicted_class_GSQ)):
    if(predicted_class_GSQ[i]==0 ):
        predicted_class.append('GALAXY')        
    elif( predicted_class_GSQ[i]==1 ):
        predicted_class.append('STAR') 
    elif( predicted_class_GSQ[i]==2 ):
        predicted_class.append('QSO') 
```

```python
partial_data['class']=predicted_class
```
