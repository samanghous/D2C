```python
import pandas as pd
```

```python
full_data=pd.read_excel('Training Dataset.xlsx')
partial_data=pd.read_excel('Testing Dataset.xlsx')
```

```python
def classify(x):
    if(x=='GALAXY'):
        return 0
    elif(x=='STAR'):
        return 1
    elif(x=='QSO'):
        return 2

full_data['class_GSQ']=full_data['class'].apply(classify)    
```

```python
# ***************WE CAN REMOVE A SMALL FRACTION FROM TRAINIG DATA FOR TESTING PURPOSE******************************
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(full_data.iloc[:,0:17],full_data['class_GSQ'],test_size=0.2)
```

```python
x_train
```

```python
#*************************************IMPORT AND APPLY RANDOM FOREST**********************************************
```

```python
from sklearn.ensemble import RandomForestClassifier
sammodel=RandomForestClassifier()
```

```python
sammodel.fit(x_train,y_train)
```

```python
sammodel.score(x_test,y_test)
```
