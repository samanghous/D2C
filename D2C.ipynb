```python
def maxleafnodes(x_train,y_train):                                              # hyperparameter tuning things
    maxval=[]                                                               # for max_leaf_nodes only
    for i in range(20,50):
        sammodel1=RandomForestClassifier(max_leaf_nodes=i,criterion='entropy',min_samples_leaf=2,min_samples_split=4)
        sammodel1.fit(x_train,y_train)
        p=sammodel1.score(x_test,y_test)
        maxval.append(i)
        maxval.append(p)
    return maxval
```

```python
#*************************HYPERPARAMETER TUNING THINGS (part 1)**************************************************
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
sammodel1=RandomForestClassifier(n_jobs=-1)

sam_hyperparam={
    'n_estimators':[30,50,60,65,70,75,80,100,120],
    'criterion':['entropy','gini'],
    'max_depth':[None,80,90,100,120,200],
    'min_samples_split':[3,4,5,6,7],
    'min_samples_leaf':[2,3,4,5],
    'bootstrap':[True,False],
    'max_features':['auto']
     
}
```

```python
def sam_hyperpm_func(itern):
    from sklearn.model_selection import RandomizedSearchCV
    sam_bestparam=RandomizedSearchCV(sammodel1,param_distributions=sam_hyperparam,n_jobs=-1,n_iter=itern,cv=5)
    sam_bestparam.fit(full_data.iloc[:,0:17],full_data['class_GSQ'])
    gh1=sam_bestparam.best_params_ 
    gh2=sam_bestparam.best_score_
    return gh1,gh2


sam1,sam2=sam_hyperpm_func(100)
```

```python
from sklearn.utils import resample                            #overpopulate minority

minority_goodsize = resample( minorityin_train, # upsample minority
                          replace=True, # sample with replacement
                          n_samples=min(len(train[train.class_GSQ==0]),len(train[train.class_GSQ==1])), # match number in majority class
                          random_state=27) # reproducible results
```
